```python
from Stock_Data_Class import StockData
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

$\textbf{0 - Data Processing}$

```python
# Utilisation de la classe
folder_path = r"data\raw"

stocks = StockData(folder_path)
stocks.infos()



apple_data = stocks.get_historical_data('AAPL')
apple_data.plot(x='Date', y='Adj Close')
```

<div align="center">

$\textbf{Correlation Analysis}$

</div>


```python
min_dates = stocks.data.groupby('Ticker')['Date'].min()
latest_start = min_dates.max()
truncated_data = stocks.data[stocks.data['Date'] >= latest_start]
pivoted_data = truncated_data.pivot(index='Date', columns='Ticker', values='Return')
correlation_matrix = pivoted_data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()

```
